<b>Objective:To predict the percentage of a student based on the no. of study hours by using simple linear regression method.Also to predict the score if the student studies for 9.25hrs/day.

<b>IMPORTING THE REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

<b>LOADING AND READING THE DATASET

```python
URL_data="http://bit.ly/w-data"  #import data from the remote link
df = pd.read_csv(URL_data)
df.head(10)
```

<b> FEATURE EXPLORATION

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

<b> VISUALIZING THE DATA

```python
plt.xlabel('No of Hours per day')
plt.ylabel('Scores of students')
plt.title('Study Hours Vs Student Scores')
plt.scatter(x='Hours',y='Scores',data= df,marker='o',color='orange')

plt.show()
```

<b> DATA PREPARATION
    

<b> 1.Split Train and Test Data

```python
x=df.iloc[:,:-1]
y=df.iloc[:,1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

  <b> 2.Applying Linear Regression

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

<b> REGRESSION LINE

```python
line= lr.coef_*x+lr.intercept_
plt.scatter(x,y)
plt.plot(x,line);
plt.show()
```

<b>CALCULATING THE ACCURACY OF THE MODEL ON TRAIN SET
    

```python
plt.scatter(x_train,y_train)
print("Train Set Score")
print(lr.score(x_train,y_train))
```

<b>CALCULATING THE ACCURACY OF THE MODEL ON TEST SET

```python
print("Test Score")
print(lr.score(x_test,y_test))
plt.scatter(x_test,y_test)
plt.plot(x_test,lr.predict(x_test),color='red')
```

<b>SCORING

```python
print(x_test)
y_pred_test=lr.predict(x_test)


#comparing actual vs predicted
new_df=pd.DataFrame({'Actual':y_test,'Predicted':y_pred_test})
new_df
```

<b> CHECKING THE TASK

```python
hours=9.25
test=np.array([hours])
test=test.reshape(-1,1)
own_pred=lr.predict(test)
print("no of hours = ",format(hours))
print("predicted scores = ",format(own_pred))
```

<b>MODEL EVALUATION

```python
print('Mean Absolute Error:', round(mean_absolute_error(y_test,y_pred_test),2))
print('Mean Squared Error:', round(mean_squared_error(y_test,y_pred_test),2))
print('R-2:', round(r2_score(y_test,y_pred_test),2))
```
